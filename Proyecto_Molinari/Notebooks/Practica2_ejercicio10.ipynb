#### Ejercicio 2: Duración de una playlist
Dada la playlist con canciones y sus duraciones en formato "m:ss", calcule la duración total de la playlist en formato Xm Ys, y encuentre la canción más larga y la más corta.
Nota: Deberá convertir las duraciones de string a valores numéricos para poder operar con ellas.

```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
total_segundos = 0
max_segundos = 0
min_segundos = 9999
cancion_mas_larga = ""
cancion_mas_corta = ""
# Recorro con un bucle la lista de diccionarios para analizar cancion por cancion y sumar cada duracion 
for cancion_actual in playlist:      # Cancion_actual representa un diccionario entero que tiene dos pares clave-valor
    min_texto, seg_texto = cancion_actual["duration"].split(":")     # Obtengo la lista de "duration" (["minutos", "segundos"] y asigno el 1er val a la 1ra variable y el 2do a la 2da
    duracion_actual_segundos = int(min_texto)*60 + int(seg_texto)     # Convierto los minutos y segundos a enteros
    total_segundos += duracion_actual_segundos            # Sumo la duracion en segundos de la cancion actual a la variable acumuladora
    if duracion_actual_segundos > max_segundos:           # Comparo y actualizo cancion mas larga y mas corta
        max_segundos = duracion_actual_segundos          
        cancion_mas_larga = cancion_actual
    if duracion_actual_segundos < min_segundos:
        min_segundos = duracion_actual_segundos
        cancion_mas_corta = cancion_actual
total_min = total_segundos // 60               # Calculo final (afuera del for): vuelvo a formato minutos, segundos
resto_segundos = total_segundos % 60
print(f"Duracion total de la playlist: {total_min}m {resto_segundos}s")
print(f"Cancion mas larga: {cancion_mas_larga['title']} ({cancion_mas_larga['duration']})")
print(f"Cancion mas corta: {cancion_mas_corta['title']} ({cancion_mas_corta['duration']})")
```

#### Ejercicio 10: Simulación de competencia de cocina y ranking
